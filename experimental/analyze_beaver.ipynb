```python
import pandas as pd
from utils import classify_question
```

```python
# Downloaded from: https://github.com/peterbaile/beaver
orig_df_dw = pd.read_json("beaver_dev_dw.json")
orig_df_nw = pd.read_json("beaver_dev_nw.json")
```

```python
orig_df = pd.concat([orig_df_dw, orig_df_nw])
```

```python
orig_df["query_type"] = orig_df.apply(
  lambda row: classify_question(row["question"], row.get("sql")), axis=1
)
print(orig_df["query_type"].value_counts())
print((orig_df["query_type"].value_counts() / len(orig_df)) * 100)
orig_df["query_type"].value_counts().plot(kind="bar")
```
